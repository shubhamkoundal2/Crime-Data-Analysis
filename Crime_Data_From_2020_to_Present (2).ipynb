# About the dataset
The dataset contains information about reported crimes, including the date and time of occurrence, crime descriptions, victim details, location coordinates, and other relevant attributes. It is a comprehensive dataset that can be analyzed to gain insights into crime patterns, demographics, and geographic distributions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Crime_Data_from_2020_to_Present.csv")
```

### **Data Cleaning**

```python
df.info()
```

```python
# Convert the "Date Rptd" column to datetime type
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'])

# Extract only the date portion from "Date Rptd" column
df['Date Rptd'] = df['Date Rptd'].dt.date

# Convert the "DATE OCC" column to datetime type
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])

# Extract only the date portion from "DATE OCC" column
df['DATE OCC'] = df['DATE OCC'].dt.date
```

```python
# Drop the "DR_NO" column
df = df.drop("DR_NO", axis=1)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# Dropping columns with high number of missing values
df.drop(['Mocodes', 'Weapon Used Cd', 'Weapon Desc', 'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street'], axis=1, inplace=True)

# Dropping rows with missing values in specific columns
df.dropna(subset=['Premis Cd', 'LAT', 'LON'], inplace=True)

# Imputing missing values with a specific value
df['Vict Sex'].fillna('Unknown', inplace=True)
df['Vict Descent'].fillna('Unknown', inplace=True)
df['Premis Desc'].fillna('Not Available', inplace=True)

# Imputing missing values with statistical measures
df['Premis Cd'].fillna(df['Premis Cd'].mean(), inplace=True)

# Verify if all null values have been handled
print(df.isnull().sum())
```

```python
# Drop rows with null values in the 'Crm Cd 1' column
df.dropna(subset=['Crm Cd 1'], inplace=True)

# Verify if null values have been dropped
print(df.isnull().sum())
```

## **Data Visualization**

## **1.What are the most common types of crimes reported in the dataset? (Analyzing the 'Crm Cd Desc' column)**

```python
crime_counts = df['Crm Cd Desc'].value_counts()
top_10_crimes = crime_counts.head(10)
```

```python
import seaborn as sns
sns.set(style="whitegrid")  # Set the style of the graph
plt.figure(figsize=(12, 6))  # Set the figure size

# Plot the bar graph
ax = sns.barplot(x=top_10_crimes.values, y=top_10_crimes.index, palette="viridis")

# Customize the graph
plt.xlabel('Count')
plt.ylabel('Crime Type')
plt.title('Top 10 Most Common Crimes Reported')
plt.xticks(fontsize=10)

# Add value labels to the bars
for i, v in enumerate(top_10_crimes.values):
    ax.text(v + 20, i, str(v), color='black', va='center')

plt.show()  # Display the graph
```

## **2.Which area has the highest number of reported crimes? (Analyzing the 'AREA NAME' column)**

```python
area_counts = df['AREA NAME'].value_counts()
sorted_areas = area_counts.sort_values(ascending=False)
sns.set(style="whitegrid")  # Set the style of the graph
plt.figure(figsize=(12, 6))  # Set the figure size
# Choose a different color palette
palette = "Set3"  # You can change this to any other palette provided by seaborn

# Plot the bar graph using the chosen palette
ax = sns.barplot(x=sorted_areas.values, y=sorted_areas.index, palette=palette)

# Customize the graph
plt.xlabel('Number of Reported Crimes')
plt.ylabel('Area')
plt.title('Number of Reported Crimes by Area')
plt.xticks(fontsize=10)

plt.show()
```

## **3.What is the age distribution of the crime victims? (Analyzing the 'Vict Age' column)**

```python
# Assuming your dataset is stored in a DataFrame called 'df'
victim_ages = df['Vict Age'].loc[df['Vict Age'] != 0].dropna()  # Exclude age of 0 and missing values

# Plotting the age distribution
plt.figure(figsize=(10, 6))
plt.hist(victim_ages, bins=20, edgecolor='black', color='blue')

plt.title('Age Distribution of Crime Victims')
plt.xlabel('Age')
plt.ylabel('Count')

plt.show()
```

## **4.What are the most frequently reported premises for crimes? (Analyzing the 'Premis Desc' column)**

```python
# Count the occurrences of each premise
premise_counts = df['Premis Desc'].value_counts().head(10)  # Select the top 10 premises

# Create a horizontal bar plot
plt.figure(figsize=(10, 6))
premise_counts.plot(kind='barh', color='skyblue')

# Customize the plot
plt.title('Top 10 Most Frequently Reported Premises for Crimes')
plt.xlabel('Frequency')
plt.ylabel('Premise Description')

plt.tight_layout()  # To ensure labels are not cut off
plt.show()
```

## **5.How do the crime statuses vary across reported incidents? (Analyzing the 'Status Desc' column)**

```python
# Count the occurrences of each crime status
status_counts = df['Status Desc'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
status_counts.plot(kind='bar', color='purple')

# Customize the plot
plt.title('Distribution of Crime Statuses')
plt.xlabel('Crime Status')
plt.ylabel('Frequency')

plt.tight_layout()  # To ensure labels are not cut off
plt.show()
```

Analyzing the dataset, it is found that the most common types of crimes reported are theft-related offenses, followed by assault and vandalism. The area with the highest number of reported crimes is Central, indicating a concentration of criminal activities in that region. The age distribution of crime victims varies, with a wide range of ages affected. The most frequently reported premises for crimes include streets, residences, and businesses. Regarding the status of reported incidents, the majority are marked as under investigation or with an arrest made. These findings highlight the prevalence of specific crimes, geographic patterns, victim demographics, and the ongoing efforts to address and resolve reported incidents.
